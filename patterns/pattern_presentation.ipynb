```python
import pandas as pd
import matplotlib.pyplot as plt
# if using a jupyter notebook
%matplotlib inline
```

```python
# definition of some official codilime colors
def rgb_to_hex(r, g, b):
    return '#{:02x}{:02x}{:02x}'.format(r, g, b)

lime = rgb_to_hex(163, 237, 0)
lagoon_main = rgb_to_hex(0, 131, 148)
turquoise_main = rgb_to_hex(68, 220, 236)
glacial_main = rgb_to_hex(76, 189, 250)
mint_main = rgb_to_hex(0, 218, 156)
```

```python
# plot selected memory consumption pattern
plt.figure(figsize=(10,4))
# pattern directory
pattern_dir = 's'
# pattern file name
pattern = 'high_start_1mT.csv'
df = pd.read_csv(f'{pattern_dir}/{pattern}')
ax = df['mem'].plot(color=[lagoon_main], label=pattern, legend=True, marker='o')
plt.title(f"Memory consumption pattern")
plt.xlabel('second')
plt.ylabel('% of maximal memory to be allocated')
plt.grid()
plt.show()
```
